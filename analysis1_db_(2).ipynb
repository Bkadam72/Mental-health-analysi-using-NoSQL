```python
import pymongo
import pandas as pd
from dtreeviz.trees import dtreeviz
from matplotlib import pyplot as plt
from sklearn import tree
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
myclient = pymongo.MongoClient("mongodb+srv://jnadar1:jnadar1DB@cluster0.gcwftxq.mongodb.net/?retryWrites=true&w=majority")
mydb = myclient["MentalHealthQ"]
mycol = mydb["covidcoll"]
```

```python
pipeline=[
    {
        '$project': {
            'any therapy or medication?': 1,
            'mental state good?': 1,
            'eating & sleeping': 1,
            'COVID-19 affected': 1,
            'sudden or huge change in life': 1,
            '30 or above': 1,
            'Employment Status': 1,
            '_id': 0
        }
    }
]
```

```python
cursor = mycol.aggregate(pipeline)
PipeList=list(cursor)
df = pd.DataFrame(PipeList)
print(df)
```

```python
d = {'Yes': 1, 'No': 0}
df['30 or above'] = df['30 or above'].map(d)
d = {'Yes': 1, 'No': 0, 'Not sure': 2}
df['COVID-19 affected'] = df['COVID-19 affected'].map(d)
df['sudden or huge change in life'] = df['sudden or huge change in life'].map(d)
d = {'Yes': 1, 'No': 0, 'Maybe': 2}
df['eating & sleeping'] = df['eating & sleeping'].map(d)
df['mental state good?'] = df['mental state good?'].map(d)
df['any therapy or medication?'] = df['any therapy or medication?'].map(d)
d = {'Employed': 1, 'Unemployed': 0, 'Self-Employed': 2, 'Student':3}
df['Employment Status'] = df['Employment Status'].map(d)
print(df)
```

```python
features = list(df.columns[:4])
features
```

```python
X = df[features]
y = df["COVID-19 affected"]

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

```python
viz = dtreeviz(clf, X, y,
                target_name="any therapy or medication?",
                feature_names=features,
                class_names=PipeList)
viz
```
